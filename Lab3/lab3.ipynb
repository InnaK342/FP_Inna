##Згенерувати набір даних за допомогою означеної функції, для бінарної класифікації. 


```python
import numpy as np 
from sklearn import datasets 
import matplotlib.pyplot as plt 
from sklearn import metrics
from sklearn.linear_model import LogisticRegression

X, y = datasets.make_classification(n_samples=1000,n_informative=1, n_features=2, n_classes=2, n_redundant=0, n_clusters_per_class=1,class_sep = 2) 
fig, ax = plt.subplots(figsize=(9, 6)) 
plt.xlabel("X0", fontsize=20) 
plt.ylabel("X1", fontsize=20) 
plt.scatter(X[:,0], X[:,1], s=50, c=y)
```

##Здійснити моделювання згенерованого набору даних та візуалізацію результатів.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
min1, max1 = X[:, 0].min()-1, X[:, 0].max()+1
min2, max2 = X[:, 1].min()-1, X[:, 1].max()+1
arr_x1 = np.arange(min1, max1, 0.01)
arr_x2 = np.arange(min2, max2, 0.01)
xx, yy = np.meshgrid(arr_x1, arr_x2)
r1, r2 = xx.flatten(), yy.flatten()
r1, r2 = r1.reshape((len(r1), 1)), r2.reshape((len(r2), 1))
grid = np.hstack((r1,r2))

model = LinearDiscriminantAnalysis()
model.fit(X, y)
pred_y = model.predict(grid)
zz = pred_y.reshape(xx.shape)
plt.contourf(xx,yy,zz,cmap='Pastel2')
plt.scatter(X[:,0], X[:,1], s=50, c=y)
plt.show()
```

##Надати значення метрик якості, приклад класифікації


```python
from sklearn.metrics import confusion_matrix

tn, fp, fn, tp = confusion_matrix(y,model.predict(X)).ravel()

tn, fp, fn, tp
```
